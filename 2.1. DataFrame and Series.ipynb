<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab:

*   List menyimpan object dengan index terurut 0, 1, 2, 3, ... suatu object di urutan tertentu tidak dapat diubah indexnya. Dictionary menyimpan object dengan index key, key memiliki value berupa object. Series dengan list mempunyai index default 0, 1, 2, 3, ..., index ini datap dimodifikasi dengan urutan yang sama. Sedangkan Series dengan dictionary mempunyai index default key dari dictionary tersebut, jika index dimodifikasi maka value dari suatu index tetap mengikuti value dari key yang sama dengan index tersebut, jika tidak ada key yang memenuhi maka diberi nilai NaN.
*   Dataframe adalah object 2D terdiri dari kolom dan baris. Sedangkan series object 1D hanya terdiri dari baris.





```python
# code here
lst = [0, 1, 2, 3]
series_lst = pd.Series(lst, index=[3, 0, 'a', 'z'])
series_lst
```

```python
dct = {1:'a', 2:2, 3:3}
series_dct = pd.Series(dct, index=[3,1,4])
series_dct
```
