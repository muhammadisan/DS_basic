<h2>Soal 1: Penggabungan Data</h2>

- Sebutkan perbedaan antara full outer join, inner join, dan left join

Jawab:  

*   Full outer join adalah union set, menggabungkan semua info pada columns tertentu yang ada pada data frame pertama **atau** data frame kedua.
*   Inner join adalah intersection set, menggabungkan info hanya jika entry pada columns yang diinginkan ada pada data frame pertama **dan** data frame kedua.
* Left join adalah menggabungkan info columns hanya yang ada pada data frame pertama (left)

<h2>Soal 3: Penggabungan Data</h2>

Lakukan jenis penggabungan data full outer join, inner join, left outer join dan right outer join pada kedua data berikut: 

```python
import pandas as pd

data1 = {
    'id' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'id' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
# join data
df_outer_join = pd.merge(df1, df2, how='outer')
df_outer_join
```

```python
df_inner_join = pd.merge(df1, df2, how='inner')
df_inner_join
```

```python
df_left_join = pd.merge(df1, df2, how='left')
df_left_join
```

```python
df_right_join = pd.merge(df1, df2, how='right')
df_right_join
```

<h2>Soal 4: Penggabungan Data berbeda label column Key</h2>

Lakukan jenis penggabungan inner join untuk 2 data yang memiliki nama column key yang berbeda ini:

```python
data1 = {
    'key_data1' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'key_data2' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
# Gabungkan kedua data tersebut
df_gabung = pd.merge(df1, df2, left_on='key_data1', right_on='key_data2', how='inner')

df_gabung
```
