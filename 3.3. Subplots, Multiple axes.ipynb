```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18,8), sharey=True)

ax[0].plot(x, tan, 'b', label='tanX')
ax[1].plot(x, sin, 'r', label='sinX')
ax[2].plot(x, cos, 'g', label='cosX')

ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(13, 9))

ax.plot(x, y)
ax.set_title('Grafik cosX')

ax1 = ax.inset_axes([-0.1,0.9, 0.3,0.3])
ax1.plot(x, y5, 'y', x, y, 'b')
ax1.set_title('Grafik CosX^5')

ax2 = ax.inset_axes([0.9,0.9, 0.3,0.3])
ax2.plot(x, y4, 'k', x, y, 'b')
ax2.set_title('Grafik CosX^4')

ax3 = ax.inset_axes([-0.1,-0.1, 0.3,0.3])
ax3.plot(x, y2, 'r', x, y, 'b')
ax3.set_title('Grafik CosX^2')

ax4 = ax.inset_axes([0.9,-0.1, 0.3,0.3])
ax4.plot(x, y3, 'g', x, y, 'b')
ax4.set_title('Grafik CosX^3')

plt.show()
```
